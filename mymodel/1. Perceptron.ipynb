```python
def sigmoid(x):
    return 1/(1+(np.exp(-x)))
```

```python
def train(x, y):
    weights = np.random.randn(2)
    bias = np.random.randn()
    for epoch in range(10000):
        for i in range(4):
            output = np.dot(x[i], weights) + bias 
            output = sigmoid(output)
            error  = y[i] - output
            weights += 0.5 * (error * x[i])
            bias    += 0.5 * error
    return weights, bias
```

```python
x = np.array([[0, 0], [0, 1], [1, 0], [1, 1]])
y = np.array([0, 1, 1, 1])
```

```python
train(x, y)
```

# Xor

```python
import numpy as np
import matplotlib.pyplot as plt

def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def sigmoid_derivative(x):
    sx = sigmoid(x)
    return sx * (1 - sx)

def error(y_hat, y):
    return 0.5 * np.mean((y - y_hat) ** 2)

def error_derivative(y_hat, y):
    return y_hat - y

def train(X, y):
    weight_input = np.random.randn(2, 2)
    bias_input = np.random.randn(1, 2)
    weight_output = np.random.randn(2, 1)
    bias_output = np.random.randn(1, 1)

    for epoch in range(10000):
        # Forward pass
        hidden_raw = np.dot(X, weight_input) + bias_input
        hidden = sigmoid(hidden_raw)

        output_raw = np.dot(hidden, weight_output) + bias_output
        output = sigmoid(output_raw)

        err = error(output, y)
        if err < 1e-4:
            break

        # Backward pass
        grad_output = error_derivative(output, y) * sigmoid_derivative(output_raw)
        grad_hidden = np.dot(grad_output, weight_output.T) * sigmoid_derivative(hidden_raw)

        # Update weights and biases
        weight_output -= 0.1 * np.dot(hidden.T, grad_output)
        bias_output -= 0.1 * np.sum(grad_output, axis=0, keepdims=True)

        weight_input -= 0.1 * np.dot(X.T, grad_hidden)
        bias_input -= 0.1 * np.sum(grad_hidden, axis=0, keepdims=True)

    return weight_input, bias_input, weight_output, bias_output

def predict(X, weight_input, bias_input, weight_output, bias_output):
    hidden = sigmoid(np.dot(X, weight_input) + bias_input)
    output = sigmoid(np.dot(hidden, weight_output) + bias_output)
    return output

def plot(X, y, weight_input, bias_input, weight_output, bias_output):
    # Plot the points
    for i in range(len(X)):
        color = 'blue' if y[i] == 0 else 'red'
        plt.scatter(X[i][0], X[i][1], color=color)

    # Decision boundary
    x_vals = np.linspace(0, 1, 100)
    y_vals = np.linspace(0, 1, 100)
    points = []
    for xv in x_vals:
        for yv in y_vals:
            pred = predict(np.array([[xv, yv]]), weight_input, bias_input, weight_output, bias_output)
            if 0.48 < pred < 0.52:  # near decision boundary
                points.append((xv, yv))
    
    if points:
        px, py = zip(*points)
        plt.plot(px, py, 'g.', label='Decision Boundary')

    plt.title("XOR Function Learned by Neural Network")
    plt.xlabel("x1")
    plt.ylabel("x2")
    plt.grid(True)
    plt.legend()
    plt.show()

# XOR dataset
X = np.array([[0, 0], [0, 1], [1, 0], [1, 1]])
y = np.array([[0], [1], [1], [0]])

# Train and plot
w_in, b_in, w_out, b_out = train(X, y)
plot(X, y, w_in, b_in, w_out, b_out)
```

```python
import numpy as np
import matplotlib.pyplot as plt

def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def sigmoid_derivative(x):
    s = sigmoid(x)
    return s * (1 - s)

def error(y_hat, y):
    return 0.5 * np.mean((y - y_hat) ** 2)

def error_derivative(y_hat, y):
    return y_hat - y

def train(X, y):
    weight_input = np.random.randn(2, 2)
    bias_input = np.random.randn(1, 2)
    weight_output = np.random.randn(2, 1)
    bias_output = np.random.randn(1, 1)

    for epoch in range(10000):
        # forward
        hidden_raw = np.dot(X, weight_input) + bias_input
        hidden = sigmoid(hidden_raw)

        output_raw = np.dot(hidden, weight_output) + bias_output
        output = sigmoid(output_raw)

        err = error(output, y)
        if err < 1e-4:
            break

        # backward
        grad_output = error_derivative(output, y) * sigmoid_derivative(output)
        grad_hidden = np.dot(grad_output, weight_output.T) * sigmoid_derivative(hidden_raw)

        weight_output -= 0.1 * np.dot(hidden.T, grad_output)
        bias_output -= 0.1 * np.sum(grad_output, axis=0, keepdims=True)

        weight_input -= 0.1 * np.dot(X.T, grad_hidden)
        bias_input -= 0.1 * np.sum(grad_hidden, axis=0, keepdims=True)

    return weight_input, bias_input, weight_output, bias_output


# XOR data
X = np.array([[0,0], [0,1], [1,0], [1,1]])
y = np.array([[0], [1], [1], [0]])

# Train
w_in, b_in, w_out, b_out = train(X, y)

# Plot scatter points
for i in range(len(X)):
    color = 'red' if y[i] == 1 else 'blue'
    plt.scatter(X[i][0], X[i][1], color=color)

# Plot decision lines from hidden neurons
x_vals = np.linspace(-0.1, 1.1, 100)
for i in range(2):  # two hidden neurons
    # Equation: w1*x + w2*y + b = 0 -> y = -(w1*x + b)/w2
    w1 = w_in[0][i]
    w2 = w_in[1][i]
    b = b_in[0][i]
    y_vals = -(w1 * x_vals + b) / w2
    plt.plot(x_vals, y_vals, label=f"Hidden neuron {i+1}")

plt.title("XOR Input Points and Hidden Neuron Lines")
plt.xlabel("x1")
plt.ylabel("x2")
plt.legend()
plt.grid(True)
plt.show()
```
